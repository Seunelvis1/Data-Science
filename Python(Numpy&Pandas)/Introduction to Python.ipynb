## Break, Continue and Pass
Break - Breaks out of the current closest enclosing loop
Continue - Goes to the top of the closest enclosing loop
Pass - Do nothing at all

```python
name = "Oluwaseun"

for i in name:
    if i == "n":
        break
    print(i)
```

```python
name = "Oluwaseun"

for i in name:
    if i == "a":
        continue
    print(i)
```

```python
for i in range(3):
    password = input("What is your password: ")
    if password == "1234":
        print("Login successfully")
        break
    print("Wrong Password")
```

### For Loop

```python
# Password  verification
correct_password = "qwerty123"
max_attempt = 5

for attempt in range(max_attempt):
    password_try = input(f"This is attempt {attempt + 1} of {max_attempt} attempts:")

    if password_try == correct_password:
        print("Acsess Granted")
        break
    else:
        print("Incorrect Password")

        if attempt < max_attempt - 1:
            print("Please try again \n")
            continue # Continue to the next  attempt
        else:
            print("You've maxed out all you attempts and you are blocked")
            break
```
