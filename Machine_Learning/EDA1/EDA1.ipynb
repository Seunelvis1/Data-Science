## Dealing with Missing Data

1. **MCAR (Missing Completely at Random):**  
   These are values that are randomly missing and do not depend on any other values.

2. **MAR (Missing at Random):**  
   These values are dependent on some additional features.

3. **MNAR (Missing Not at Random):**  
   There is a reason behind why these values are missing.

---

### Methods for Handling Missing Data

Several methods exist for imputing missing data, including:

- **Central Tendency Imputation:**  
  Replacing missing values with the **mean**, **median**, or **mode** of the variable.

- **Regression Imputation:**  
  Using other variables in the dataset to predict the missing values.

- **Multiple Imputation:**  
  Generating multiple estimates for missing data to account for uncertainty and variability.

```python
# Convert the date column into a pandas datetime object
df["Date"] = pd.to_datetime(df["Date"])

df.info()
```

```python
# Extract the Year, Month and Quarter
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['month_name'] = df['Date'].dt.month_name()
df['quarter'] = df['Date'].dt.quarter

df.head(2)
```

```python
# Group Customer_Age

def age_group(x):
    if x <= 25:
        return "<= Young Adult"
    elif x <= 40:
        return "25-40 Adult"
    elif x <= 50:
        return "41-50 Old Adult"
    else:
        return ">=51 Elders"

# Applying function to the data
df['age_group'] = df["Customer_Age"]. apply(age_group)
df.head(2)
```

```python
# Cost, Revenue and Profit calculation

df["cost"] = df["Quantity"]* df["Unit_Cost"]
df["revenue"] = df["Quantity"]* df["Unit_Price"]
df["profit"] = df["revenue"]- df["cost"]

df.head(2)
```

```python
# Profit or loss
def PorL(x):
    if x >= 0:
        return "Profit"
    else:
        return "Loss"

df['Profit_label'] = df["profit"].apply(PorL)

df.head()
```
